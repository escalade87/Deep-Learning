```
import numpy as np
```

Head is 1, Tail is 0

```
def coinFlip():    
    #perform the binomial distribution (returns 0 or 1)    
    result = np.random.binomial(1,.5)
    #return flip to be added to numpy array    
    return result
```

```
def average(lst): 
    return sum(lst) / len(lst) 
```

```
def repeatCoinFlip(n, type):
  fullResults = []

  for i in range(0, n): 
    #initiate array
    tempResults = []

    #perform flips at required probability set above
    while True:    
        tempResults.append(coinFlip())
        length = len(tempResults)

        if length > 2:
          if tempResults[length-3] == 1:
            if tempResults[length-2] == 0:
              if type == 1:
                if tempResults[length-1] == 1:
                  break
              elif type == 2:
                if tempResults[length-1] == 0:
                  break
    #print results
    #print("Tails = 0, Heads = 1: ", tempResults)
    #print("Number of coin toss : ",len(tempResults))
    fullResults.append(len(tempResults))

  #print("Result : ", fullResults)
  print("Average : ", average(fullResults))
  return average(fullResults)
```

```
def repeatExperiment(n, type):
  endResults = []
  for i in range(0, n):
    endResults.append(repeatCoinFlip(n,type))
  
  print("Average Experiment : ", average(endResults))
```

type 1 = HTH


```
n = 1000

repeatExperiment(n,1)
```


type 2 = HTT

```
repeatExperiment(n,2)
```

```
from pandas import DataFrame

Data = {'height': [170,165,174,169,155,172,166,168],
        'weight': [60,55,75,67,49,63,58,61],
        'grade': [4.1,3.0,2.8,2.9,3.1,3.6,3.7,4.0]
        }

df = DataFrame(Data,columns=['height','weight','grade'])

cov = df.cov()
print("Covariance :\n")
print(cov)

corrMatrix = df.corr()
print("\nCorrelation Matrix :\n")
print (corrMatrix)
```
